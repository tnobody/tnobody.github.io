```python
from collections import Counter

def get_lines(file: str):
    with open(file, 'r') as file:
        lines = file.readlines()
    return lines

matrix = [list(word.strip()) for word in get_lines('words.txt')]
matrix_t = [list(word) for word in zip(*matrix)]

char_counts = [Counter(word) for word in matrix_t]
```

```python
import pandas as pd

pd.DataFrame([counter.most_common(100) for counter in char_counts]).T
```

```python
def word_is_allowed(word, ignore_chars=[], fixed_chars="     ", required_chars="     "):
    no_ignored_chars = all([c not in ignore_chars for c in word])
    matches_fixed_chars = all([c == word[i] for (i,c) in enumerate(fixed_chars) if c is not ' '])
    required_but_not_at = all([c is not word[i] for (i, c) in enumerate(required_chars) if c is not ' '])
    contains_requires = all([c in word for c in required_chars if c is not ' '])
    return no_ignored_chars and matches_fixed_chars and required_but_not_at and contains_requires


def get_word_score(word):
   return sum([
      char_counts[i][c] 
      for (i, c) in enumerate(word)
   ])


def get_best_words(ignore_chars=[], required_chars="     ", fixed_chars="     "):   
   df = pd.DataFrame(
       [
           [get_word_score(word), "".join(word)]
         for word in matrix
         if word_is_allowed(word, ignore_chars=ignore_chars, required_chars=required_chars, fixed_chars=fixed_chars)
        ],
      columns=['Score', 'Word']
   )
   return df.sort_values(by=['Score'], ascending=False)

get_best_words(
   ignore_chars=list('soresnty'),
   required_chars='ca  h',
   fixed_chars="  ac "
)
```
